```python
## Importing dependencies and libraries required

import pandas as pd
```

```python
## Importing Required DataSets of Anime and the ratings of previously watched audience
## Merging on the basis of 'MAL_id' and dropping unneeded data for collaborative filtering

ratings = pd.read_csv('rating_complete.csv')
details = pd.read_csv('anime_with_synopsis.csv')
complete_ds = pd.merge(details,ratings)
anime_cf_ds = pd.merge(details,ratings).drop({'Score','Genres','sypnopsis'},axis=1)
complete_ds.head()
```

```python
details.head()
```

```python
##Required dataset to perform Item-Item Collaborative filtering

anime_cf_ds.head()
```

```python
## Depicting rating of all users for all anime 

user_ratings = anime_cf_ds.pivot_table(index=['user_id'],columns=['Name'],values='rating')
user_ratings.head()
```

```python
## Dropping all the anime which have less than 10 viewers and by putting unseens rating as 0

user_ratings = user_ratings.dropna(thresh=20,axis=1).fillna(0)
```

```python
user_ratings.head()
```

```python
## Building recommendation on pearson model, to give optimum execution time is to efficiency ratio in comparison to other models

item_similarity_df = user_ratings.corr(method='pearson')
item_similarity_df.head(50)
```

```python
## Declaring function to recommend anime based on data recieved using pearson model

def get_similar_anime(Name,user_ratings):
    similar_score = item_similarity_df[Name]*(user_ratings-5)
    similar_score = similar_score.sort_values(ascending=False)
    
    return similar_score    
```

```python
## Taking user input by providing his/her ratings on already watched anime

anime_lover = [("Naruto",8),("Hyouka",10),("Fruits Basket",8),("Durarara!!",7)]
similar_anime = pd.DataFrame()
for anime,rating in anime_lover:
    similar_anime = similar_anime.append(get_similar_anime(anime,rating),ignore_index = True)

similar_anime.head(10)
```

```python
## Recommending anime in descending order

similar_anime.sum().sort_values(ascending=False).head(20)
```

```python
recomend_anime = similar_anime.sum().sort_values(ascending=False)
recomend_anime.head(20)
```
