```python
## Importing dependencies and libraries required

import numpy as np
import pandas as pd
```

```python
## Importing Required DataSets of Hollywood movie

movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
## Seeking data included in base dataset

movies.head(1)
```

```python
## Seeking cast and crew data based of different movies

credits.head(1)
```

```python
## forming an array data collection of all the string values of crew column in credit dataset

credits.head(1)['crew'].values
```

```python
## Merging movies and credit dataset to get a better dataset

movies.merge(credits,on='title')
```

```python
## Substituting the merged dataset into initial movies dataset

movies = movies.merge(credits,on='title')
```

```python
## Verifying the row cross column value of the merged dataset

movies.shape
```

```python
## Keeping only the genre, id, keywords, title, overview, cast and crew to provide recommendations

movies = movies[['movie_id','title','overview','genres','keywords','runtime','release_date',"vote_average",'cast','crew']]
```

```python
## Viewing the processed dataset

movies.head()
```

```python
## Verfying row cross column number

movies.shape
```

```python
## Checking for null i.e. not available values

movies.isnull().sum()
```

```python
## Removing NaN values

movies.dropna(inplace=True)
```

```python
## Removing duplicated values

movies.duplicated().sum()
```

```python
## Seeking the genres of movie at key [0] of the array 

movies.iloc[0].genres
```

```python
## Changing string value to array for combining all values together

import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
## Applying the function declared above

movies['genres']= movies['genres'].apply(convert)
```

```python
## Result of applied function

movies.head()
```

```python
## Repeating the same step for other string variables

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
## Combining first name and last name of cast to avoid error due to same first name or sur name of two or more cast

def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break            
    return L
```

```python
## Applying the above function to cast column

movies['cast']=movies['cast'].apply(convert3)
```

```python
## Result of above converting function

movies.head()
```

```python
## Seeking crew data of movie to find the format of data present

movies['crew'][10]
```

```python
## Sorting director of the movie out of the whole crew

def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break            
    return L
```

```python
## Applying fetch_director function

movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
## Adding overview section in array form

movies['overview'][0]
```

```python
## Furnishing the array overview

movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
## Adding genre section in array form by ommiting space

movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
## Declaring 'tags' column with combination of 'overview','genres','keywords','cast' & 'crew'

movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
## new dataframe with selective info as much need

new_movie_Df= movies[['movie_id','title','overview','genres','runtime','release_date','vote_average','cast','crew','tags']]
```

```python
new_movie_Df['tags']=new_movie_Df['tags'].apply(lambda x:" ".join(x))
```

```python
new_movie_Df['overview']=new_movie_Df['overview'].apply(lambda x:" ".join(x))
```

```python
new_movie_Df['genres']=new_movie_Df['genres'].apply(lambda x:" ".join(x))
```

```python
new_movie_Df['cast']=new_movie_Df['cast'].apply(lambda x:" ".join(x))
```

```python
new_movie_Df['crew']=new_movie_Df['crew'].apply(lambda x:" ".join(x))
```

```python
## Final processed dataset after returning the formed array form to its original data type

new_movie_Df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
## Function to omit repetative or similar words from overview

def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
## Applying function to decrease similar words in tags

new_movie_Df['tags']=new_movie_Df['tags'].apply(stem)
```

```python
new_movie_Df['tags'][0]
```

```python
## Making 'tags' lowercase

new_movie_Df['tags']=new_movie_Df['tags'].apply(lambda x:x.lower())
```

```python
new_movie_Df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
## Converting 'tags' to a vector to apply cosine_similarity to deduce most recommendable movie

vectors=cv.fit_transform(new_movie_Df['tags']).toarray()
```

```python
## output vectors

vectors
```

```python
cv.get_feature_names()
```

```python
## Reducing similar words throughout 'tags'

stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction culture clash future space war space colony society space travel futuristic romance space alien tribe alien planet cgi marine soldier battle love affair anti war power relations mind and soul 3d sam worthington zoe saldana sigourney weaver james cameron')
```

```python
## Importing library to apply cosine_similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
## Applying cosine_similarity to the vector processed from 'tags' column

cosine_similarity(vectors)
```

```python
## Declaring similarity score

similarity=cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
## Defining and declaring function to find recommendations based on closest available vector based using cosine_similarity

def recommend(movie):
    movie_index = new_movie_Df[new_movie_Df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:11]
    
    for i in movies_list:
        print(new_movie_Df.iloc[i[0]].title)
    
```

```python
## Implementation of recommend function

recommend('Avengers: Age of Ultron')
```
